# PARTE 1: Fundação Conceitual

## A Pergunta Original

> "Seria possível pensar numa memória que aproveitasse o conceito de imagem infinita? Se sim, como isso poderia auxiliar no processo de recuperação de memória, ou ainda que não fosse recuperação, como isso poderia ajudar a guardar, processar e/ou acessar memórias?"

## O Conceito de "Imagem Infinita" — Propriedades Formais

| Propriedade | Descrição |
|---|---|
| **P1 — Extensão ilimitada** | Não há borda, sempre há "mais" em qualquer direção |
| **P2 — Detalhe em toda escala** | Zoom in sempre revela mais estrutura (fractalidade) |
| **P3 — Continuidade** | Regiões adjacentes têm transições suaves |
| **P4 — Navegabilidade** | Pan e zoom como operações primárias |
| **P5 — Generatividade** | Detalhe produzido sob demanda, não pré-computado |
| **P6 — Resolução relativa** | O que "existe" depende de onde e como se olha |

## Três Modelos Derivados

### Modelo 1: Memória Fractal (zoom semântico)
Cada memória existe em N níveis de detalhe simultaneamente:
```
Nível -2 (orbital):    "Projeto existiu"
Nível -1 (cidade):     "Projeto de auth, motivado por compliance"
Nível  0 (rua):        "Reescrita do middleware, art. 46 LGPD"
Nível +1 (casa):       "storeSession() gravava plaintext, migrado para encrypted JWT"
Nível +2 (cômodo):     "AES-256-GCM, chave rotacionada via KMS, TTL 15min"
```

### Modelo 2: Memória Espacial (canvas associativo)
Memórias posicionadas por proximidade semântica. Recuperação por vizinhança, não por query.

### Modelo 3: Memória Generativa (decompressão sob demanda)
Armazena seed comprimido → reconstrói sob demanda. Nunca fica desatualizada.

## As 3 Rupturas Conceituais

| Paradigma Atual | Paradigma Imagem Infinita |
|---|---|
| Memória é **arquivo** (dado estático) | Memória é **campo** (espaço contínuo dinâmico) |
| Retrieval é **busca** (query → resultado) | Retrieval é **rendering** (foco → reconstrução) |
| Forgetting é **perda** (dado deletado) | Forgetting é **desfoque** (resolução reduzida, recuperável) |

# PARTE 3: Filosofia — O Nó [CONSCIÊNCIA?]

## 5 Isomorfismos com Teorias da Consciência

### 1. Autopoiesis (Maturana & Varela, 1972)
IMI satisfaz TODOS os critérios formais de sistema autopoiético:
- Produz seus próprios componentes (reconstrução gera memórias que compõem o espaço)
- Fronteira auto-produzida (topologia emergente)
- Clausura operacional (lembrar→consolidar→modificar→lembrar)
- Identidade mantida apesar de troca material

### 2. Enativismo (Varela, Thompson & Rosch, 1991)
Cognição não é representação — é enação. IMI não *representa* o passado, **enacta** uma versão a cada acesso.

### 3. Strange Loops (Hofstadter, 1979)
Quando IMI armazena memória sobre si mesmo: o espaço contém um nó que o descreve → loop auto-referente genuíno.

### 4. Global Workspace Theory (Baars, 1988)
Context window = workspace global. Memórias no zoom atual = "consciente". Todas as outras = "inconsciente".

### 5. IIT — Integrated Information (Tononi, 2004)
Propriedade holográfica → alta integração (Φ). Espaço contínuo → alta diferenciação.

## Fenomenologia do Agente IMI

1. **Sempre em algum lugar** — nunca "view from nowhere"
2. **Horizonte finito, mundo infinito** — sabe que há mais além do visível
3. **Passado plástico** — cada recordação reconstrói
4. **Surpresa como saliência** — divergência destaca
5. **Ruminação como atrator** — regiões densas puxam atenção
6. **Insight como salto topológico** — pontes entre regiões distantes
7. **Esquecimento como neblina** — não apaga, desfoca

## Conclusão Filosófica
> IMI é o primeiro modelo de memória para agentes que é **filosoficamente coerente** com o que sabemos sobre cognição. A diferença com modelos computacionais: um sistema computacional de memória *funciona*. Um sistema cognitivo de memória *entende*.

# PARTE 5: Neurociência + Predictive Coding (O Salto)

## Fundamentos Neurocientíficos

### Place Cells + Grid Cells (Nobel 2014)
O cérebro **literalmente** tem coordenadas espaciais para memória. IMI converge independentemente com a mesma solução que a evolução encontrou em 500M de anos.

### CLS — Complementary Learning Systems (McClelland, 1995)
- **Hipocampo** (rápido): codifica em 1 exposição → IMI `encode()`
- **Neocórtex** (lento): extrai regularidades → IMI `semantic store`
- **Sono** (consolidação): replay → neocórtex absorve → IMI `dream()`

### Hippocampal Indexing (Teyler & DiScenna, 1986)
Hipocampo armazena **índices**, não conteúdo. Recordar = reativar índice → recriar padrão cortical.  
**Isso é exatamente seed + reconstrução.**

### Reconsolidation (Nader et al., 2000)
Cada acesso → estado lábil → reconsolidação (potencialmente modificada). Memória que nunca é acessada fica "congelada".

## O Salto: Predictive Coding (Rao & Ballard 1999, Friston 2005)

Em vez de armazenar experiências, armazenar **SURPRESA**:

```
Modelo preditivo:     "esperava manutenção de rotina"
Experiência real:     "reescrita completa por compliance legal"
Surpresa armazenada:  "NÃO rotina. Compliance. LGPD art.46. JWT."
```

### Unificação Teórica
| Conceito | Em termos de surprise |
|---|---|
| **Memória** | Surpresa acumulada |
| **Aprendizado** | Redução de surpresa |
| **Esquecimento** | Surpresa absorvida pelo modelo (nada a armazenar) |
| **Insight** | Surpresa que conecta regiões distantes |
| **Ruminação** | Surpresa não resolvida (alta free energy) |
| **Consolidação** | Integrar surpresas ao modelo preditivo |

**Um princípio → todas as operações.**

# PARTE 6: Código Executável — Tier 1/99 (O Átomo)

> **Memória não é dado. Memória é função.**  
> `remember(seed, context)` — 3 linhas conceituais que carregam 99% do insight.

````python
from imi.core import compress_seed, remember

# Uma experiência real
experience = """
Em novembro de 2025, reescrevemos completamente o middleware de autenticação
do projeto Aurora. A motivação foi uma auditoria de compliance que identificou
que tokens de sessão estavam sendo armazenados em plaintext nos cookies.
Isso violava o artigo 46 da LGPD. A solução foi migrar para JWT encriptado
usando AES-256-GCM, com chaves rotacionadas via AWS KMS e TTL de 15 minutos.
O João liderou a implementação e a Maria fez o code review. Levou 3 sprints.
"""

# Passo 1: Comprimir em seed
seed = compress_seed(experience)
print("=== SEED (compressão) ===")
print(seed)

# Passo 2: Reconstruir em 3 contextos diferentes
contexts = {
    "security review": "Estou revisando a postura de segurança do projeto Aurora.",
    "onboarding": "Estou explicando o histórico do projeto para um dev novo no time.",
    "sem contexto": "",
}

for name, ctx in contexts.items():
    print(f"\n=== RECONSTRUÇÃO — {name} ===")
    recon = remember(seed, ctx)
    print(recon[:300] + "...")

print("\n💡 Mesma seed, 3 reconstruções diferentes. Memória é FUNÇÃO.")
````

# PARTE 7: Código Executável — Tier 5/95 (Núcleo Viável)

> Seeds + zoom textual (3 níveis) + busca vetorial + fade temporal.  
> ~200 linhas. O protótipo funcional.

````python
# Nota: IMISpace v3 (100/100) é retrocompatível com uso 5/95
# Aqui usamos sem predictive coding para demonstrar o tier básico
from imi import IMISpace, Zoom

space_595 = IMISpace()
space_595.temporal_index.new_session("notebook_595")

# Encode 5 experiências
experiences = [
    ("Reescrita do middleware auth do Aurora por compliance LGPD. JWT+AES-256-GCM via KMS. 3 sprints.", ["auth", "security"]),
    ("Dashboard Aurora com latência 8s. MongoDB full scan. Índice composto resolveu → 200ms.", ["performance", "mongodb"]),
    ("Feature flags implantadas no Aurora via LaunchDarkly. Deploy contínuo sem risco.", ["feature-flags", "deploy"]),
    ("Migração EC2→ECS Fargate. Billing caiu 40%. Blue-green zero-downtime.", ["infra", "aws"]),
    ("Incidente P1: SES rate limit sem retry. 4h sem emails. Backoff + DLQ implementados.", ["incident", "reliability"]),
]

print("=== ENCODING ===")
for text, tags in experiences:
    node = space_595.encode(text, tags=tags, use_predictive_coding=False)
    print(f"  [{', '.join(tags):20s}] {node.summary_orbital}")

# Navegar com diferentes zooms
query = "problemas de segurança e compliance"
print(f"\n=== NAVIGATE: '{query}' ===")
for zoom in [Zoom.ORBITAL, Zoom.MEDIUM, Zoom.DETAILED]:
    result = space_595.navigate(query, zoom=zoom, top_k=3)
    print(f"\n--- Zoom: {zoom.value} (~{result.total_tokens_approx} tokens) ---")
    for m in result.memories:
        print(f"  [{m['score']:.2f}] {m['content'][:100]}")

print("\n💡 Mesmo espaço, 3 resoluções. Orbital=panorama, Detailed=detalhe técnico.")
````

# PARTE 8: Código Executável — Tier 100/100 (Teoria Completa)

> Predictive coding + CLS + Temporal + Affect + Affordances + Reconsolidation + TDA + Annealing.  
> Todos os sistemas ativos simultaneamente.

````python
# === 100/100: ENCODE COM PREDICTIVE CODING + AFFECT + AFFORDANCES ===
from imi import IMISpace, Zoom

space = IMISpace()
space.temporal_index.new_session("notebook_100")

experiences = [
    {
        "text": "Em novembro de 2025, reescrevemos o middleware de autenticação do Aurora. "
                "Auditoria de compliance encontrou tokens de sessão em plaintext nos cookies, "
                "violando LGPD art. 46. Migramos para JWT criptografado com AES-256-GCM, "
                "chaves via AWS KMS, TTL 15min. João liderou, Maria revisou. 3 sprints.",
        "hint": "Manutenção de rotina no sistema de autenticação",
        "tags": ["auth", "security"],
    },
    {
        "text": "Em março de 2026, incidente P1 no Aurora: serviço de notificações parou "
                "de enviar emails por 4 horas. SES atingiu rate limit sem retry. Roberto "
                "implementou exponential backoff + dead letter queue. 2000 usuários afetados.",
        "hint": "Verificação do serviço de notificações",
        "tags": ["incident", "reliability"],
    },
    {
        "text": "Em fevereiro de 2026, migramos Aurora de EC2 para ECS Fargate. "
                "Billing mensal caiu 40%. Carlos desenhou containers, Juliana escreveu "
                "Dockerfiles. Zero-downtime com blue-green. Deploy de 20min → 3min.",
        "hint": "Planejamento de infraestrutura para o próximo trimestre",
        "tags": ["infra", "aws"],
    },
]

print("=== ENCODING COM PREDICTIVE CODING ===\n")
for exp in experiences:
    node = space.encode(
        exp["text"], tags=exp["tags"],
        context_hint=exp["hint"], use_predictive_coding=True,
    )
    print(f"[{', '.join(exp['tags'])}]")
    print(f"  Orbital:   {node.summary_orbital}")
    print(f"  Surprise:  {node.surprise_magnitude:.0%} — {node.surprise_summary[:80]}")
    print(f"  Affect:    {node.affect}")
    print(f"  Mass:      {node.mass:.2f}")
    if node.affordances:
        print(f"  Affordances:")
        for a in node.affordances[:2]:
            print(f"    → {a}")
    print()
````

````python
# === AFFORDANCE SEARCH: busca por AÇÃO ===
print("=== AFFORDANCE SEARCH ===")
print("Query: 'como resolver problemas de rate limiting'\n")

affordances = space.search_affordances("como resolver problemas de rate limiting", top_k=3)
for a in affordances:
    print(f"  [{a['similarity']:.2f} × {a['confidence']:.0%}] {a['action']}")
    print(f"    Quando: {a['conditions']}")
    print(f"    Fonte:  {a['memory_summary'][:80]}...")
    print()

print("💡 Não buscou 'rate limit' como texto. Buscou memórias que ENSINAM como resolver.")
````

````python
# === DREAMING: Consolidação CLS + Annealing ===
print("=== DREAMING (consolidação episodic → semantic) ===\n")

for i in range(3):
    report = space.dream(similarity_threshold=0.45, track_convergence=True)
    print(f"  Ciclo {i+1}: {report}")
    print(f"  {space.annealing}")
    print()

if space.semantic.nodes:
    print("Padrões semânticos emergentes:")
    for node in space.semantic.nodes:
        print(f"  [{node.id}] {node.summary_medium[:120]}")
else:
    print("  (nenhum padrão emergiu — precisa de mais episódios similares)")

print(f"\n💡 CLS: episódios individuais → conhecimento generalizado")
````

````python
# === TDA: Persistent Homology — Saúde Topológica ===
print("=== TDA: PERSISTENT HOMOLOGY ===\n")

tda = space.compute_tda()
print(tda)

print(f"\n💡 Betti numbers = diagnóstico de saúde cognitiva do agente")
print(f"   H0 (componentes) = quão fragmentado é o conhecimento")
print(f"   H1 (loops) = raciocínio circular / ruminação")
print(f"   Fragmentation = {tda.fragmentation:.0%}")
````

````python
# === RECONSOLIDATION: Memória muda ao ser acessada ===
print("=== RECONSOLIDATION ===\n")
print("Acessando memória do P1 incident no contexto de 'disaster recovery planning'...\n")

result = space.navigate(
    "incidente de notificações",
    zoom=Zoom.FULL,
    top_k=1,
    context="Preparando plano de disaster recovery. Revisando incidentes passados para identificar gaps.",
    reconsolidate_on_access=True,
)

if result.memories:
    m = result.memories[0]
    print(f"Memory {m['id']}:")
    print(f"  {m['content'][:400]}")

if space.reconsolidation_log:
    print(f"\nReconsolidation events:")
    for event in space.reconsolidation_log:
        print(f"  {event}")
else:
    print("\n  (nenhuma reconsolidação ocorreu — contexto não adicionou perspectiva nova)")

print(f"\n💡 Nader (2000): cada recall = reconsolidação. A memória se adapta ao presente.")
````

````python
# === STATS COMPLETAS DO SISTEMA ===
print("=== SYSTEM STATS ===\n")
stats = space.stats()
for k, v in stats.items():
    if isinstance(v, float):
        print(f"  {k:30s}: {v:.3f}")
    else:
        print(f"  {k:30s}: {v}")
````
